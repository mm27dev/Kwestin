```python
import sqlite3
from google.colab import drive
try:
  drive.mount('/content/drive')
except:
  pass
con = sqlite3.connect('/content/drive/MyDrive/SharedDatabase/question.db')
cur = con.cursor()
file_path= '/content/drive/MyDrive/SharedDatabase/QuestionDatabase.txt'

try:
  cur.execute("CREATE TABLE questions( Question)")
  con.commit() # Commit table creation if successful
except sqlite3.OperationalError:
  print(".")

name = input("What is your Name: ")
questionbank = []
def again():
  print("Any more Questions? (Yes/No):")
  global another
  another = input()
  if another.lower() == "yes":
      question()
  else:
    # When done, select all questions, close connection
    print("\n--- All Questions in Database ---")
    rows = cur.execute("SELECT Question FROM questions").fetchall()
    writeinfo= ''
    for row in rows:
          writeinfo = writeinfo + row[0]+ "\n"
    with open(file_path, 'w') as f:
            f.write(writeinfo)


    con.close() # Close the connection
    print("Database connection closed.")

def question():
  print("what is your question:")
  user_question = input() # Renamed to avoid conflict with function name
  questionbank.append(user_question)
  cur.execute("INSERT INTO questions VALUES(?)", (name + "-" + user_question,))
  con.commit() # Commit after each insert
  again()

# Start the process
question()
```
